 How to use a function in python


```python
def plus_two(x):   #def defines the function, plus_two is the function name, x is the varible that is used internally. (this does not overwrite the variable x if it is used elswhere in the code)
    return x+2
```

```python
plus_two(5)
```

Lets define a function for the fibonacci sequence:

```python
def fibonacci (N):
    L = []
    a, b = 0, 1
    while len(L) < N:
      a, b = b, a +b
      L.append (a)

    return L
```

```python
fibonacci (10)
```

You can add the functionality of default argument values where if nothing is provided the function can still run without error using those values. For example:

```python
def fibonacci_2 (N, a = 0, b = 1):
    L = []
    while len(L) < N:
      a, b = b, a +b
      L.append (a)

    return L
```

```python
fibonacci_2(10)
```

```python
fibonacci_2(10, a = 1 , b = 3)
```

For functions with an unknown amount of inputs, you need the function to be flexible, in this case you can use *args and **kwargs

```python
def catch_all(*args, **kwargs):
    print("args =", args)
    print("kwargs = ", kwargs)
```

```python
catch_all(1, 2, 3, a=4, b=5)
catch_all('a', keyword=2)
```

Python uses variables as pointers (not the case for all languages which may use them as containers) since the variables are used as pointers they point to an object (string, float, int, etc.) if you set two variables equal and then change what one vaible is pointing at, it will be chnanged for both.

```python
x_point = [1,2,3]
y_point = x_point
print(y_point)
```

```python
x_point.append(4)
print(y_point)
```

Notice how changing the variable x_point changed the variable y_point
